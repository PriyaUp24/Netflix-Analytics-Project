```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#IMPORT LIBRARIES
#Pandas for data manipulation, NumPy for numerical operations, Matplotlib for visualization.
```

```python
#LOAD DATA
df = pd.read_csv('../data/netflix_titles.csv')
```

```python
df.head()
```

```python
#INITIAL DATA CHECK
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
#- Checked dataset size, structure, and data types
#- Identified missing values and categorical variables
```

```python
#DATA CLEANING
# Remove duplicate records
df = df.drop_duplicates()

# Handle missing values
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.isnull().sum()
#Missing categorical values were replaced with ‘Unknown’ to preserve row integrity.
```

```python
#CONTENT TYPE DISTRIBUTION
df['type'].value_counts()
```

```python
#This analysis shows the distribution between Movies and TV Shows on Netflix.
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
#VISUALIZATION
```

```python
df['date_added'].dt.year.value_counts().sort_index()
```

```python
df['date_added'].dt.year.value_counts().sort_index().plot()
plt.title('Content Added Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10)
#Identifies countries contributing the most content to Netflix.
```

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10)
#GENRE ANALYSIS
```

```python
df['rating'].value_counts()
#RATINGS ANALYSIS
```

```python
## Key Insights
- Movies dominate Netflix’s content library
- Rapid growth in content after 2015
- USA and India are top content producers
- Drama and International genres are most common
- Majority content targets mature and teen audiences
```
